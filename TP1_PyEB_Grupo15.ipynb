# Estimación de coeficientes de una regresión lineal

Un problema muy común en la estadística es el de estimación. En este
caso, tendremos una variable objetivo, $Y$, que se desea estimar a
partir de $r$ variables observables, $X_1, X_2, \ldots, X_r$. Un modelo
de regresión lineal para este problema será de la forma

$$
Y = \beta_0+\beta_1X_1 + \beta_2X_2+\ldots+\beta_rX_r + \varepsilon,
$$

donde $\varepsilon$ es un término de error que podemos suponer sigue una
distribución Normal de media 0 y varianza desconocida
$\sigma_\varepsilon^2$.

El objetivo del problema de regresión lineal resulta hallar los valores
óptimos para $\beta_0, \beta_1,\ldots,\beta_r$ a partir de $n$
observaciones del vector aleatorio $(Y,X_1,\ldots,X_r)$. Una forma de
hacerlo es a partir del estimador de máxima verosimilitud.

Para ello debemos observar que dada una observación
$\boldsymbol{x} = (x_1,\ldots, x_r)$, la distribución condicional de $Y$
para a ser también Normal:

$$
Y|\boldsymbol{X}=\boldsymbol{x} \sim \mathcal{N}(\beta_0+\beta_1x_1+\ldots+\beta_rx_r, \sigma_\varepsilon²)
$$

Partimos de una variable aleatoria Y definida por la siguiente ecuación

$$
Y = \beta_0 + \beta_1X_1 +\varepsilon

Además, tenemos un vector de $n$

Partimos con una observación

para a ser también Normal con la siguiente distribución:

$$
Y|\boldsymbol{X}=\boldsymbol{x} \sim \mathcal{N}(\beta_0+\beta_1x_1, \sigma_\varepsilon²)
$$

Y el objetivo del problema está en encontrar los valores óptimos de $\beta_0$ y $\beta_1 $ que ajusten mejor los datos que ingresemos al modelo(esto lo haremos en el siguiente ejercicio, con el dataset Income).

##A) Búsqueda del Estimador de Máxima Verosimilitud

Como nuestra variable Y condicionada a X tiene distribución Normal con los parámetros definidos anteriormente, tenemos que su función de densidad es la siguiente

$$
f_{(\beta_0, \beta_1)} (y) = \frac{1}{\sqrt{2\pi} \sigma} e^{- \frac{(y_i - (\beta_0 + \beta_1 x_i))^2}{2 \sigma ^2}}
$$

Observamos que el soporte de la densidad de la variable Y NO depende de $\theta $(Esto nos viene bien a la hora de querer encontrar el mejor estimador, utilizando la función ln para alivianar las cuentas). Ahora buscamos el Estimador de Máxima Verosimilitud para $\beta_0$ y para $\beta_1$, quedándonos con una función de dos variables.

$$
L(\beta_0, \beta_1) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi} \sigma} e^{- \frac{(y_i - (\beta_0 + \beta_1 x_i))^2}{2 \sigma ^2}}
$$

Desarrollando un poco la expresión

$$
L(\beta_0, \beta_1) = {(\frac{1}{\sqrt{2\pi} \sigma})^n} e^{\sum_{i=1}^{n}- \frac{(y_i - (\beta_0 + \beta_1 x_i))^2}{2 \sigma ^2}}
$$

Como dijimos antes, el soporte de la función de densidad NO depende de $\theta$, por lo que podemos aplicar $ln$ a ambos lados de la expresión para maximizar la función de verosimilitud(el logaritmo es una función monótona creciente).

$$
ln(L(\beta_0, \beta_1)) = n. ln{(\frac{1}{\sqrt{2\pi} \sigma})} + \sum_{i=1}^{n}- \frac{(y_i - (\beta_0 + \beta_1 x_i))^2}{2 \sigma ^2}
$$

&nbsp;

$$
ln(L(\beta_0, \beta_1)) = n. ln{(\frac{1}{\sqrt{2\pi} \sigma})} + \sum_{i=1}^{n}- \frac{(y_i - \beta_0 - \beta_1 x_i)^2}{2 \sigma ^2}
$$

Teniendo esta expresión para la función de verosimilitud, vamos a maximizarla, esto es, derivando parcialmente con respecto a $\beta_0$ y con respecto a $\beta_1$, e igualaremos a 0 a ambas derivadas, quedando un sistema de ecuaciones del cual encontraremos los valores de $\beta_0$ y $\beta_1$ que mejor se ajusten a los datos.

$$
\frac{\partial ln(L(\beta_0, \beta_1))}{\partial \beta_0} = - \frac{1}{2 \sigma^2} \sum_{i=1}^{n} 2 (y_i - \beta_0 - \beta_1 x_i). (-1)
$$
&nbsp;

$$
\frac{\partial ln(L(\beta_0, \beta_1))}{\partial \beta_1} = - \frac{1}{2 \sigma^2} \sum_{i=1}^{n} 2 (y_i - \beta_0 - \beta_1 x_i). (-x_i)
$$

Igualamos a 0 ambas expresiones, las simplificamos y armamos el sistema de ecuaciones
&nbsp;

$$
\left\{
    \begin{array}{l}
        0 = \frac{1}{\sigma^2} \sum_{i=1}^{n}(y_i - \beta_0 - \beta_1 x_i) &\\
        0 = \frac{1}{\sigma^2} \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i). x_i
    \end{array}
\right.
$$
&nbsp;

Lo desarrollamos un poco más.
&nbsp;

$$
\left\{
    \begin{array}{l}
        0 = \sum_{i=1}^{n}(y_i - \beta_0 - \beta_1 x_i) &\\
        0 = \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i). x_i
    \end{array}
\right.
$$
&nbsp;

Desarrollamos la primera ecuación del sistema buscando $\beta_0$.
$$
0 = \sum_{i=1}^{n}(y_i - \beta_0 - \beta_1 x_i) = \sum_{i=1}^{n}y_i - n\beta_0 - \beta_1\sum_{i=1}^{n}x_i
$$

$$
n\beta_0 = \sum_{i=1}^{n}y_i - \beta_1\sum_{i=1}^{n}x_i
$$

$$
\beta_0 = \frac{1}{n}(\sum_{i=1}^{n}y_i - \beta_1\sum_{i=1}^{n}x_i)
$$

$$
\boxed{\beta_0 = \bar{y} - \beta_1\bar{x}}
$$

Ahora, desarrollamos la segunda ecuacion del sistema buscando $\beta_1$.

$$
0 = \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i). x_i = \sum_{i=1}^{n}y_ix_i - \beta_0\sum_{i=1}^{n}x_i - \beta_1\sum_{i=1}^{n}x_i^2
$$

$$
\beta_1\sum_{i=1}^{n}x_i^2 = \sum_{i=1}^{n}y_ix_i - \beta_0\sum_{i=1}^{n}x_i
$$

Reemplazamos $\beta_0$ por lo que obtuvimos al resolver la primera ecuación del sistema.

$$
\beta_1\sum_{i=1}^{n}x_i^2 = \sum_{i=1}^{n}y_ix_i - (\bar{y} - \beta_1\bar{x})\sum_{i=1}^{n}x_i
$$

$$
\beta_1\sum_{i=1}^{n}x_i^2 = \sum_{i=1}^{n}y_ix_i - \bar{y}\sum_{i=1}^{n}x_i + \beta_1\bar{x}\sum_{i=1}^{n}x_i
$$

$$
\beta_1\sum_{i=1}^{n}x_i^2 - \beta_1\bar{x}\sum_{i=1}^{n}x_i = \sum_{i=1}^{n}y_ix_i - \bar{y}\sum_{i=1}^{n}x_i
$$

$$
\beta_1(\sum_{i=1}^{n}x_i^2 - \bar{x}\sum_{i=1}^{n}x_i) = \sum_{i=1}^{n}y_ix_i - \bar{y}\sum_{i=1}^{n}x_i
$$

$$
\boxed{\beta_1 = \frac{\sum_{i=1}^{n}y_ix_i - \bar{y}\sum_{i=1}^{n}x_i}{\sum_{i=1}^{n}x_i^2 - \bar{x}\sum_{i=1}^{n}x_i}}
$$

### B) Estimación de $\beta_0$ y $\beta_1$ usando R

Insertamos los datos correspondientes a las ecuaciones obtenidas en el ejercicio 1 del TP.

```r
y_media = sum(XY$happiness)/nrow(XY)
x_media = sum(XY$income)/nrow(XY)
```

```r
beta1 = (sum(XY$income * XY$happiness)-y_media*sum(XY$income))/(sum(XY$income^2)-x_media*sum(XY$income))
beta0 = y_media - beta1*x_media
```

Graficamos nuevamente los datos y superponemos la recta estimada a partir de
los valores de $\hat\beta_0$ y $\hat\beta_1$ hallados.

```r
plot(XY)
abline(a=beta0, b=beta1, col='aquamarine3', lw=3)
```
